#1. ユークリッド幾何学の公理について
- 公理について、調べてください。初等幾何学における定理はもともとは、いろんな土木工事などの作業等での経験的知識が集大成されたものだと思います。それらが、五つの公理から演繹を用いて導かれるというのがギリシャ人の凄いところですね。これにはいろんな意味で重要なことがあります。しかしここでは、ユークリッド幾何をこれらの公理から出発しなくて、$R^n$に標準的な内積を入れたユークリッド空間、自体ですべてが尽くされる、したがって、本来初等幾何は不要の産物です。 しかし、定理の内容によっては、初頭幾何的に証明したほうが楽な場合も多々あります。従って、ユークリッド幾何は有限次元実ベクトル空間に内積が入ったものと定義しておいて、補助的に初頭幾何的な証明を使う立場をとります。
- ここで、逆に非ユークリッド幾何について考えてみましょう。これは５つの公理のうち一つだけ変えたものです。詳しいことは自分で調べて欲しいですが、球面のモデルと双曲面ないしはポアンカレの上半平面のモデルがあります。ここで注意して欲しいのはこれは定曲率の正または負の曲面になっており、ここでいう直線とは、測地線になっており空間が曲がっているものになってきます。つまりユークリッド幾何の公理における直線とか平面はある意味で無定義概念で、その意味を変えれば、一つ公理変えることに成立したわけです。
- またユークリッド幾何における公理は長い間絶対的心理と考えられてきました、したがって非ユークリッド幾何を考えることは、タブーだったわけで、虚数が存在しないという神学的（？）な証明があったくらいですから、迫害されたりしたわけです。現代では公理は単なる仮定、約束事にすぎません。しかし公理から出発して演繹を用いていろんな主張を導くという今日の数学の在り方を大きく決定づけたのはユークリッドの公理でしょうね。
- 前述のとおり、ユークリッドの公理における「直線」や「交わる」などの用語は曖昧さを持っています。現代の数学では、集合論ないしは圏論の言葉を基礎としています。しかし例えば集合を定義せよといえばゲーデルの不完全性定理に直面します。それはペンディングとしても、線形空間として定義した方がはるかにすっきりすると思います。
